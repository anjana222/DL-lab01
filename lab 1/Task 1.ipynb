```python
import numpy as np
import matplotlib.pyplot as plt

# Define the scale parameter for the exponential distribution
scale = 1.0

# Create a random 100000x1 array from the exponential distribution
array_100000x1_exp = np.random.exponential(scale=scale, size=(100000, 1))

# Create random arrays for uniform and normal distributions
array_100000x1_uniform = np.random.uniform(low=0.0, high=2.0, size=(100000, 1))
array_100000x1_normal = np.random.normal(loc=0.0, scale=1.0, size=(100000, 1))

# Plot histograms
plt.figure(figsize=(18, 5))

# Exponential distribution histogram
plt.subplot(1, 3, 1)
plt.hist(array_100000x1_exp, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.axis([0, 10, 0, 0.5])  # Adjusted axis limits for better visualization
plt.legend(loc="upper right")
plt.title("Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Density")

# Uniform distribution histogram
plt.subplot(1, 3, 2)
plt.hist(array_100000x1_uniform, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.axis([0, 2, 0, 1.5])  # Adjusted axis limits for better visualization
plt.legend(loc="upper right")
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")

# Normal distribution histogram
plt.subplot(1, 3, 3)
plt.hist(array_100000x1_normal, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.axis([-5, 5, 0, 0.5])  # Adjusted axis limits for better visualization
plt.legend(loc="upper right")
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range and step for X and Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z
Z = X**2 + Y**2

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Plot of Z = X^2 + Y^2')

# Show the plot
plt.show()
```

```python
# Pandas for managing datasets
import pandas as pd
# Matplotlib for additional customization
from matplotlib import pyplot as plt
# Seaborn for plotting and styling
import seaborn as sns

# Ensure inline plots
%matplotlib inline

# Load the dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Define the features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Check if all required features are in the DataFrame
missing_features = [feature for feature in features if feature not in df.columns]
if missing_features:
    raise ValueError(f"Missing features: {missing_features}")

# Select relevant features
data = df[features]

# Calculate Pearson and Spearman correlations
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')

# Plot Pearson Correlation Heatmap
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=.5)
plt.title('Pearson Correlation Heatmap')

# Plot Spearman Rank Correlation Heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=.5)
plt.title('Spearman Rank Correlation Heatmap')

plt.tight_layout()
plt.show()
```
